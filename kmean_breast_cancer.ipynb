```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer as lbc
from sklearn.cluster import KMeans
```

```python
data=lbc()
df = pd.DataFrame(data['data'],columns=data['feature_names'])
df['target']=data['target']
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
pca_res = PCA(2).fit_transform(df[data["feature_names"]])
```

```python
plt.scatter(pca_res[:, 0], pca_res[:, 1], c=df["target"], cmap="coolwarm")
```

```python
model=KMeans(2)
cluster_labels = model.fit_predict(df[data['feature_names']])
```

```python
cluster_labels
```

```python
plt.scatter(df['mean radius'],df['mean concavity'],c=df['target'])
```

```python
plt.scatter(df['mean radius'],df['mean concavity'],c=cluster_labels)
```
